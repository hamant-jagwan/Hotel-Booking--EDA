# DATA CLEANING


```python

# This user define function will give the type,count of null and non null values as well as null ratio
def hotelbookinginfo():
  temp=pd.DataFrame(index=hb_df.columns)
  temp["datatype"]=hb_df.dtypes
  temp["not null values"]=hb_df.count()
  temp["null value"]=hb_df.isnull().sum()
  temp["% of the null value"]=hb_df.isnull().mean()*100
  temp["unique count"]=hb_df.nunique() # Return Series with number of distinct elements. Can ignore NaN values.
  return temp
hotelbookinginfo()
     
```

In the above observation:

* children column has 4 null values which contain only 0.0033 of that data.
* country column has 488 null values which contain 0.40% of that data.
* agent columm has 16340 null values which contain  13 % of that data.
* Company column has 112593 null values which contain 94 % of that data so,after giving a look what we can do with it if it doesn't give value we directly drop it.  

Lets first deal with the columns which contains lesser number of NaN values. By going through the NaN values, we must come up with a way to replace them with non NaN values or we need to come up with a reason for having NaN.


##1). `children` : There are a total of 4 NaN values in this column.

```python

# The rows containing NaN values in the children column
hb_df[hb_df["children"].isnull()]  # detect missing values in the given series object
```

```python
# Finding the different values the 'Android Ver' column takes
hb_df["children"].value_counts() #Return a Series containing counts of unique values
```

Since the NaN values in the children column cannot be replaced by any particular value, and, since there are only 4 rows which contain NaN values in this column, which accounts to less than 0.003% of the total rows in the given dataset, it can be be dropped.

```python
# dropping rows corresponding to the to the NaN values in the 'children' column.
hb_df=hb_df[hb_df['children'].notna()] #detects existing/ non-missing values in the dataframe.
# Shape of the updated dataframe
hb_df.shape
     
```

```python
hb_df[hb_df["children"].isnull()]
```

```python
# to get the outlier in children column
sns.boxplot(x = hb_df['children'])
plt.title('outlier finding')
 
```

```python
hb_df.drop(hb_df[hb_df['children'] == 10].index,inplace   = True)
```
